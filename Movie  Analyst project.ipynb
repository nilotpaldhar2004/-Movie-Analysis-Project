```python
# Sort the DataFrame by gross earnings in descending order and save the result
sorted_df = df.sort_values(by=['gross'], ascending=False)

# Display the top 10 highest-grossing movies
print("Top 10 Highest-Grossing Movies:")
top_movies = sorted_df.head(10)

# Create a more informative display with selected columns
# Adjust the columns based on what's available in your dataset
# Common movie dataset columns include: title, director, year, gross, budget, rating, etc.
display_columns = ['title', 'director', 'year', 'gross', 'budget'] 
display_columns = [col for col in display_columns if col in sorted_df.columns]

# Display the formatted results
print(top_movies[display_columns].reset_index(drop=True))

# Calculate return on investment if budget column exists
if 'budget' in sorted_df.columns and 'gross' in sorted_df.columns:
    # Add ROI column (Return on Investment)
    sorted_df['roi'] = (sorted_df['gross'] - sorted_df['budget']) / sorted_df['budget'] * 100
    
    # Show top 10 movies by ROI (most profitable)
    print("\nTop 10 Most Profitable Movies (by ROI):")
    top_roi = sorted_df.sort_values(by=['roi'], ascending=False).head(10)
    roi_columns = display_columns + ['roi'] if 'roi' not in display_columns else display_columns
    print(top_roi[roi_columns].reset_index(drop=True))

# Show summary statistics for top movies vs. overall dataset
print("\nComparison of Top 10 Movies vs. Overall Dataset:")
if 'gross' in sorted_df.columns:
    print(f"Average gross for top 10 movies: ${top_movies['gross'].mean():,.2f}")
    print(f"Average gross for all movies: ${sorted_df['gross'].mean():,.2f}")
    print(f"Highest gross: ${sorted_df['gross'].max():,.2f}")
    print(f"Lowest gross: ${sorted_df['gross'].min():,.2f}")

# Return the sorted DataFrame for further use
sorted_df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (if not already loaded)
# df = pd.read_csv("your_dataset.csv")  

# Factorize only categorical columns
df_factorized = df.copy()
for col in df.select_dtypes(include=['object']).columns:  # Select only categorical columns
    df_factorized[col], _ = df[col].factorize()  # Convert categorical values to numerical labels

# Compute Correlation Matrix
correlation_matrix = df_factorized.corr(method='pearson')

# Set up the figure size for better readability
plt.figure(figsize=(12, 8))

# Generate heatmap with better color and formatting
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title and labels for better understanding
plt.title("Correlation Matrix for Movies", fontsize=14, fontweight="bold")
plt.xlabel("Movie Features", fontsize=12)
plt.ylabel("Movie Features", fontsize=12)

# Show the plot
plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

# Convert matrix into list format
corr_pairs = correlation_mat.unstack()

# Sort pairs based on absolute correlation value (ignoring self-correlations)
sorted_corr_pairs = corr_pairs.abs().sort_values(ascending=False)

# Display the top 10 strongest correlations (excluding self-correlations)
print(sorted_corr_pairs[sorted_corr_pairs < 1].head(10))
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

# Exclude self-correlations (which are always 1.00)
filtered_sorted_pairs = sorted_pairs[sorted_pairs < 1]

# Print the top 10 strongest correlations
print(filtered_sorted_pairs.tail(10))  # Last 10 values (highest correlations)
```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# Filter strong correlations (excluding self-correlations)
strong_pairs = sorted_pairs[(abs(sorted_pairs) > 0.5) & (abs(sorted_pairs) < 1)]

# Display the top 10 highest correlations
print(strong_pairs.tail(10))  # Shows last 10 values (highest correlations)
```
